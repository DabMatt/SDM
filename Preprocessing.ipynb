## Import all necessary libraries

```python
import numpy as np
import random as rd
import cv2
import glob
from skimage import io
from matplotlib import pyplot as plt
```

## Pre-processing the 'neg' images (not containing humans)
What we need here is to create, for each image from DataToPreProcess, 10 random patches of size 64x128, and to store them in Datasets

```python
origin = "DataToPreProcess\\Train\\Neg\\"
# This is the path to the folder where we find the original images
img_mask = origin+"*.*"    #The * are unknown characters
img_names = glob.glob(img_mask) # This command allows us to look at all the files having a path similar to img_mask
destination = "Datasets\\Train\\Neg\\"
# This is the destination folder
for path in img_names:
    print("processing %s..." % path,) #We look at each image in the folder
    img = cv2.imread(path)
    N =len(img[0])             #We have images of different sizes
    M = len(img)
    if N<=64 or M<=128:       #But we need to only consider the ones with bigger dimensions than 64 and 128
          print("error : image too small")
    else:
          for i in range(10):
            y = rd.randint(0,M-128)      #We create 10 patches of size 64x128 for each image
            x = rd.randint(0,N-64)       #The patches are random but we make sure they can exist in this image
            patch = img[y:y+128,x:x+64,:]
            temp = path.replace(origin,"") #This is the name of the original image (path minus path of directory)
            index = temp.find(".") #We find the index of the dot in that name
            name = temp[:index] +"_patch_"+str(i)+"_"+temp[index:] 
            #We add the patch number to the name of the original image create the name of the patch
            b = cv2.imwrite(destination+name,patch)
            # Then we write the new image at its destination
            print(b) # Writes True if the image has been saved, False else
print("Done !")
```

## Pre-processing the 'pos' images (containing humans)
Here we need, for each image, to delete columns and rows of pixels to build an image of size 64x128
It is important to delete the same number of rows from the top as from the bottom of the image (and to delete the same number of columns from the right as from the left of the image)
This way we make sure to preserve the center of the image where the human is located


```python
origin = "DataToPreProcess\\Train\\Pos\\"
# This is the path to the folder where we find the original images
img_mask = origin+"*.*"    #The * are unknown characters
img_names = glob.glob(img_mask) # This command allows us to look at all the files having a path similar to img_mask
print(img_names)
destination = "Datasets\\Train\\Pos\\"
# This is the destination folder
for path in img_names:
    print("processing %s..." % path,) #We look at each image in the folder
    img = cv2.imread(path)
    M = len(img)  #We may not have images of different sizes
    N =len(img[0])  #We use this so we won't have to rewrite a new algorithm for other image sizes
    print(N,M)
    if N<=64 or M<=128:       #But we need to only consider the ones with bigger dimensions than 64 and 128
          print("error : image too small")
    else:
        B = (M-128)//2  #We compute the difference between the actual dimensions of the image and the dimension 64x128
        D = (N-64)//2   #We delete (difference//2) on each side of the image to get the new one of dimensions 64x128
        patch = img[B:M-B,D:N-D,:]
        name = path.replace(origin,"") #This is the name of the original image (path minus path of directory)
        b = cv2.imwrite(destination+name,patch)
        # We write the new image at its destination (we didn't change the name of the image)
        print(b) # Writes True if the image has been saved, False else
print("Done!")
```

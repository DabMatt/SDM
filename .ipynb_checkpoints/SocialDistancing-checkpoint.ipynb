## Display Distances between people not respecting social distancing

## Import all necessary libraries

```python
import numpy as np
import dill
import cv2
import matplotlib.pyplot as plt
```

## Import necessary saved files

```python
img = cv2.imread("Pedestrians\\Registered_01.jpg")
img2 = cv2.imread("Pedestrians\\ScreenShotVideoResized.jpg")
```

```python
file1 = open('Saved\\Homography_matrix','rb')
H = dill.load(file1)
file1.close()

file2 = open('Saved\\computeDistance','rb')
computeDistance = dill.load(file2)
file2.close()

file3 = open('Saved\\human_detection','rb')
human_detection = dill.load(file3)
file3.close()

file4 = open('Saved\\HumansLocationsSession','rb')
dill.load_session(file4)
file4.close()
```

```python
def SocialDistancing(img):
    [result, coord, rect] = human_detection(img)
    for k in range(len(coord)):
        p1 = coord[k]
        p1_tuple = (p1[0], p1[1])
        for j in range(len(coord)-k-1):
            p2 = coord[j]
            p2_tuple = (p2[0], p2[1])
            distance = computeDistance(p1, p2, H)
            result2 = result.copy()
            if distance > 182.88:
                # We draw red rectangles on img
                cv2.rectangle(result2, (rect[k, 0], rect[k, 1]), (rect[k, 2], rect[k, 3]), (255, 0, 0), 2) 
                cv2.rectangle(result2, (rect[j, 0], rect[j, 1]), (rect[j, 2], rect[j, 3]), (255, 0, 0), 2) 
                # We draw a red line on img between those two points
                cv2.line(result2, p1_tuple, p2_tuple, (255, 0, 0), 2)
    return result2
```

```python
img_sdm = SocialDistancing(img)

plt.axis("off")
plt.imshow(cv2.cvtColor(img_sdm, cv2.COLOR_BGR2RGB))
plt.show()

cv2.imwrite("Results\\Registered_01_SDM.jpg", img_sdm)
```

```python
img2_sdm = SocialDistancing(img2)

plt.axis("off")
plt.imshow(cv2.cvtColor(img2_sdm, cv2.COLOR_BGR2RGB))
plt.show()

cv2.imwrite("Results\\ScreenShotVideoResized_SDM.jpg", img2_sdm)
```
